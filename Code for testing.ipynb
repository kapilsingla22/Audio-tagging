# Loading X test

```python
import numpy as np
import os
dirx=os.listdir('../input/assign-2-test-data/test/X')
dirx.sort()
X=np.zeros([2500,64,1000])
X=X.astype('float32')
for i in range(2500):
    x=np.load('../input/assign-2-test-data/test/X/'+dirx[i])
    X[i]=np.reshape(x,[x.shape[1],x.shape[2]])
X = X.reshape(2500,64,1000,1)
```

# Loading model

```python
model=load_model('../input/model-2-changedweights/model.h5')
```

# Predicting labels

```python
y_pred=model.predict(X)
y_predn=y_pred.round()
```

```python
label=['Alarm_bell_ringing', 
    'Blender', 
     'Cat', 
   'Dishes', 
    'Dog',
    'Electric_shaver_toothbrush', 
     'Frying', 
    'Running_water',
    'Speech', 
    'Vacuum_cleaner']
```

```python
k1=["" for h in range(2500)]

for j in range(2500):
    k=""
    for i in range(10):
       if(y_predn[j,i]==1):
        k=k+","+(label[i])
    k1[j]=k
    
```

```python
y_s=pd.read_csv('../input/test-new-csv-file/testnew.csv')
y_s['fileid']=dirx
y_s['prediction']=k1
y_s.to_csv('new_final.csv', index=False)
```
